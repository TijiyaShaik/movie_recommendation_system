**Stemming**

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
stem(new_df['tags'][0])
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```
